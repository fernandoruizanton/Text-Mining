```python
jobs = bs.findAll('div',{'class':'jobwrap','type':'tuple'})
```

```python
jobs_df = pd.DataFrame(columns=['title','company','location'])
for job in jobs:
    title=job.find('span',{'class':'title_in'}).text
    company = job.find('a',{'class':'jtxt orange'}).get('title')
    location = job.find('div',{'class':'jtxt jico ico1'}).text
    experience = job.find('div',{'class':'jtxt jico ico2'}).text
    posted_date = job.find('div',{'class':'job_optitem ico7','itemprop':'datePosted'}).text
    key_skill = job.find('div',{'class':'joblnk serachjoblnk'}).findAll('div')[4].get('title')
    curr_job = {'title':title,
               'company':company,
               'location':location,
               'experience':experience,
               'posted_date':posted_date,
               'key_skill':key_skill}
    jobs_df= jobs_df.append(curr_job, ignore_index=True)
```

```python
jobs_df.to_csv('monster.csv',index=False)
```

```python
def parse_url(url):
    page = requests.get(url).text
    bs = BeautifulSoup(page,'html.parser')
    return (bs)
def get_parents(bs):
    return(bs.findAll('div',{'class':'jobwrap','type':'tuple'}))
def get_attributes(job):
    title=job.find('span',{'class':'title_in'}).text
    company = job.find('a',{'class':'jtxt orange'}).get('title')
    location = job.find('div',{'class':'jtxt jico ico1'}).text
    experience = job.find('div',{'class':'jtxt jico ico2'}).text
    posted_date = job.find('div',{'class':'job_optitem ico7','itemprop':'datePosted'}).text
    key_skill = job.find('div',{'class':'joblnk serachjoblnk'}).findAll('div')[4].get('title')
    curr_job = {'title':title,
               'company':company,
               'location':location,
               'experience':experience,
               'posted_date':posted_date,
               'key_skill':key_skill}
    return curr_job

jobs_df = pd.DataFrame(columns=['title','company','location','experience','posted_date','key_skill'])

for i in range(1,10):
    url ='https://www.monsterindia.com/data-science-%djobs.html'% i
    print(url)
    bs = parse_url(url)
    parents = get_parents(bs)
    for job in parents:
        try:
            jobs_df =job_df.append(get_attributes(job),ignore_index=True)
        except:
            pass
        
```
